```python
import json

ccs_report_path = "/lustre/scratch126/casm/team274sb/lr26/PacBio_raw/1_A01/m64094e_230126_154129.primary_qc/ccs.report.json"

# Open and load the JSON file
with open(ccs_report_path, "r") as f:
    ccs_data = json.load(f)

# Extract key attributes
hifi_reads = next(attr["value"] for attr in ccs_data["attributes"] if attr["id"] == "ccs2.number_of_ccs_reads")
hifi_yield = next(attr["value"] for attr in ccs_data["attributes"] if attr["id"] == "ccs2.total_number_of_ccs_bases")
hifi_read_length = next(attr["value"] for attr in ccs_data["attributes"] if attr["id"] == "ccs2.mean_ccs_readlength")
hifi_quality = next(attr["value"] for attr in ccs_data["attributes"] if attr["id"] == "ccs2.median_qv")
hifi_npasses = next(attr["value"] for attr in ccs_data["attributes"] if attr["id"] == "ccs2.mean_npasses")

# Print key metrics
print(f"HiFi Reads: {hifi_reads}")
print(f"HiFi Yield: {hifi_yield} bp")
print(f"Mean HiFi Read Length: {hifi_read_length} bp")
print(f"Median HiFi Read Quality: Q{hifi_quality}")
print(f"Mean Number of Passes: {hifi_npasses}")
```

```python
import json

control_report_path = "/lustre/scratch126/casm/team274sb/lr26/PacBio_raw/1_A01/m64094e_230126_154129.primary_qc/control.report.json"

# Open and load the JSON file
with open(control_report_path, "r") as f:
    data = json.load(f)

# Extract control read metrics
control_reads = next(attr["value"] for attr in data["attributes"] if attr["id"] == "control.reads_n")
control_readlength = next(attr["value"] for attr in data["attributes"] if attr["id"] == "control.readlength_mean")
control_concordance_mean = next(attr["value"] for attr in data["attributes"] if attr["id"] == "control.concordance_mean")
control_concordance_mode = next(attr["value"] for attr in data["attributes"] if attr["id"] == "control.concordance_mode")

# Print results
print(f"Number of Control Reads: {control_reads}")
print(f"Control Read Length (Mean): {control_readlength} bp")
print(f"Control Read Concordance (Mean): {control_concordance_mean:.3f}")
print(f"Control Read Concordance (Mode): {control_concordance_mode:.3f}")
```
